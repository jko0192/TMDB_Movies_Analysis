# TMDB Movies Analysis

## Data Summary
> For this project, the TMDB Movies dataset will be explored and analyzed. The dataset consists of movies from year 1960 to 2015 with attributes including revenue, cast, budget, popularity, and more. The dataset has 10,866 rows with 21 columns

## Questions
><b>What variables influence and predict high revenues in movies?</b>

><b>Which genres are most popular from year to year?</b>

## Gathering Data

```python
# Import python packages and csv file
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline

df = pd.read_csv(r'tmdb-movies.csv')
```

- There is one duplicate row in the dataset
- The genres variable has values with multiple genres separated by a pipe that should be split into multiple columns
- The cast variable has values with multiple casts separated by a pipe that should be split into multiple columns
- The production_companies variable has values with multiple production companies separated by a pipe that should be split into multiple columns

### Data Type Issues
- The release_date variable is an object data type but it should be a datetime data type
- The genres variable is an object data type but it should be a categorical data type

```python
print('The total number of rows in the dataset: {}'.format(df.shape[0]))
print('The total number of columns in the dataset: {}'.format(df.shape[1]))
print('The total number of duplicate rows in the dataset: {}'.format(df.duplicated().sum()))
```

```python
print('Variables with missing values!')
nulls = df.isnull().sum()
nulls[nulls > 0]
```

```python
#Re-arrange the columns so the categorical variables are to the left and the numeric variables are to the right as preference
cat_variables = ['release_year', 'release_date', 'original_title', 'genres', 'cast', 'director', 'production_companies']
num_variables = ['budget','budget_adj','revenue','revenue_adj','runtime','popularity','vote_count','vote_average']
variables = cat_variables + num_variables
df = df[variables]
```

## Cleaning Data

### Fix Tidy Issues
- There is one duplicate row in the dataset
- The genres variable has values with multiple genres separated by a pipe that should be split into multiple columns
- The cast variable has values with multiple casts separated by a pipe that should be split into multiple columns
- The production_companies variable has values with multiple production companies separated by a pipe that should be split into multiple columns

```python
#Remove duplicates and split the variables genres, cast, and production_companies into their own columns
df = df[~df.duplicated()].reset_index(drop=True)
df['genre'] = df.genres.str.split('|', expand=True)[0]
df['actor'] = df.cast.str.split('|', expand=True)[0]
df['company'] = df.production_companies.str.split('|', expand=True)[0]
df = df.drop(columns=['genres', 'cast', 'production_companies'])
```

```python
#Re-arrange the columns so the categorical variables are to the left and the numeric variables are to the right as preference
column_order = list(df.columns.values[0:4]) + list(df.columns.values[-3:]) + list(df.columns.values[4:-3])
df = df[column_order]
```

### Fix Data Type Issues
- The release_date variable is an object data type but it should be a datetime data type
- The genres variable is an object data type but it should be a categorical data type

```python
# Convert variable release_date into datetime data type and variable genre into categorical data type
df.release_date = df.release_date.astype('datetime64')
df.genre = df.genre.astype('category')
df.info()
```

## Univariate Exploration

```python
# Explore the variables budget, budget_adj, revenue, and revenue_adj visually as histograms for interesting patterns
# Perform log transformations on the x-axis to create log normal distributions
# Set tick labels and axis labels correctly
variables = ['budget', 'budget_adj', 'revenue', 'revenue_adj']
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=[20,15])
axes = axes.flatten()

for index, variable in enumerate(variables):
    min_value = 0
    max_value = np.log10(df[variable].max())
    bin_size = (max_value - min_value) / 50
    bins = 10 ** np.arange(min_value, max_value + bin_size, bin_size)
    color = sb.color_palette(palette='pastel')[0]
    label = ' '.join([string.capitalize() for string in variable.split('_')]).replace('Adj', '(adjusted)')
    title = ' '.join([string.capitalize() for string in variable.split('_')]).replace('Adj', '(adjusted)')
    tick_locations = [1e1, 1e3, 1e5, 1e7, 1e9]
    tick_labels = [10, '1000', '100K', '1M', '1B']
    
    axes[index].hist(data=df, x=variable, bins=bins, color=color)
    axes[index].set_xscale('log')
    axes[index].set_xticks(tick_locations)
    axes[index].set_xticklabels(tick_labels)
    axes[index].set_xlabel('{} (log10)'.format(label), fontsize=12)
    axes[index].set_ylabel('Frequency', fontsize=12)
    axes[index].set_title('{} By Frequency'.format(title))
```

### Frequency Distributions For Budget And Revenue
> The frequency distributions for budget, budget adjusted, revenue, and revenue adjusted are unimodal after a log transformation on the x-axis. The frequency distributions are left-skewed given the enormous range of values, with the bulk of the distribution near the upper end. The shape of these frequency distributions look similar to each other indicating strong correlation

```python
# Explore the variables runtime, popularity, vote_count, and vote_average visually as histograms for interesting patterns
# Perform log transformations on the x-axis to create log normal distributions
#Set appropriate tick labels and axis labels
variables = ['runtime', 'popularity', 'vote_count', 'vote_average']
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=[20,15])
axes = axes.flatten()

for index, variable in enumerate(variables):
    color = sb.color_palette(palette='pastel')[1]
    label = ' '.join([string.capitalize() for string in variable.split('_')])
    title = ' '.join([string.capitalize() for string in variable.split('_')])
    
    if variable == 'vote_average':
        min_value = df[variable].min()
        max_value = df[variable].max()
        bin_size = (max_value - min_value) / 20
        bins = np.arange(min_value, max_value + bin_size, bin_size)
        
        axes[index].hist(data=df, x=variable, bins=bins, color=color)
        axes[index].set_xlabel('{}'.format(label), fontsize=12)
    else:
        min_value = 0
        max_value = np.log10(df[variable].max())
        bin_size = (max_value - min_value) / 50
        bins = 10 ** np.arange(min_value, max_value + bin_size, bin_size)
        tick_locations = [1e0, 1e1, 1e2, 1e3]
        tick_labels = [1, 10, 100, 1000]
        
        axes[index].hist(data=df, x=variable, bins=bins, color=color)
        axes[index].set_xscale('log')        
        axes[index].set_xticks(tick_locations)
        axes[index].set_xticklabels(tick_labels)
        axes[index].set_xlabel('{} (log10)'.format(label), fontsize=12)
        
    axes[index].set_ylabel('Frequency', fontsize=12)
    axes[index].set_title('{} By Frequency'.format(title))
```

### Frequency Distributions For Runtime, Popularity, And Votes
> The frequency distributions on runtime, after a log transformation on the x-axis, and vote average are unimodal. The frequency distribution for vote average looks like a normal distribution, but there could be slight left-skewness. 

> The frequency distributions on popularity and vote count, after a log transformation on the x-axis, appear right-skewed with the bulk of the distribution near the lower end

```python
#Explore the variable genre as a bar chart and order the types of genre by frequency
fig = plt.figure(figsize=[10,8])
color = sb.color_palette(palette='pastel')[2]
order = df.genre.value_counts().index

sb.countplot(data=df, y='genre', color=color, order=order)
plt.xlabel('Frequency', fontsize=12)
plt.ylabel('Genre', fontsize=12)
plt.title('Genre By Frequency');
```

### Frequency Distribution For Genre
> The frequency distribution for genre shows drama with the highest frequency and foreign with the lowest frequency. The frequency distribution shows drama, comedy, and action as the 3 biggest genres of where the bulk of movies are categorized

## Bivariate Exploration

```python
#Explore the pairwise correlations between numeric variables and identify strong positive correlation values with revenue
variables = ['budget', 'budget_adj', 'revenue', 'revenue_adj','runtime', 'popularity', 'vote_count', 'vote_average']
fig = plt.figure(figsize=[15,8])

h = sb.heatmap(df[variables].corr(), annot=True, fmt='.3f', cmap='mako', vmin=0)
h.set_title('Pairwise Correlation On Numeric Attributes')
h.tick_params(axis='both', labelsize=12);
```

```python
# Explore the pairwise correlations between numeric variables identified previously by the heat map 
# as potential strong predictors for revenue with scatter plots
variables = ['budget', 'budget_adj', 'revenue', 'revenue_adj', 'popularity', 'vote_count']
color = sb.color_palette(palette='pastel')[3]

grid = sb.PairGrid(data=df, vars=variables)
grid = grid.map_diag(plt.hist, bins=20, color=color)
grid = grid.map_offdiag(plt.scatter, color=color);
```

```python
# Explore the strong numeric variables individually with revenue as scatter plots
# Perform log transformations on variables popularity and vote_count to create a log normal distribution if possible
# Set appropriate tick labels and axis labels
variables = ['budget', 'budget_adj', 'revenue', 'revenue_adj', 'popularity', 'vote_count']
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=[20,12])
axes = axes.flatten()

for index, variable in enumerate(variables):
    color = sb.color_palette(palette='pastel')[3]
    label = ' '.join([string.capitalize() for string in variable.split('_')]).replace('Adj', '(adjusted)')
    title = ' '.join([string.capitalize() for string in variable.split('_')]).replace('Adj', '(adjusted)')
    
    if variable == 'popularity' or variable == 'vote_count':
        axes[index].scatter(data=df, x=variable, y='revenue', alpha=0.5, color=color)
        axes[index].set_xscale('log')
        axes[index].tick_params(axis='x', labelsize=14)
        axes[index].set_xlabel('{} (log10)'.format(label), fontsize=14)
    else:
        axes[index].scatter(data=df, x=variable, y='revenue', alpha=0.5, color=color)
        axes[index].tick_params(axis='x', labelsize=14)
        axes[index].set_xlabel('{}'.format(label), fontsize=14)
    
    axes[index].set_ylabel('Revenue', fontsize=14)
    axes[index].set_title('Correlation Between {} And Revenue'.format(title), fontsize=14)
fig.tight_layout()
```

### Pairwise Correlation Between Numeric Attributes

> The pairwise correlation between numeric attributes in the heat map shows several strong predictors for revenue. These strong predictors are budget, budget adjusted, revenue adjusted, popularity, and vote count. These predictors have strong positive correlation values of 0.67 or greater, indicating strength.

> The scatter plot correlations between revenue and predictor variables budget and budget adjusted show increasing strength when shown visually with a positive direction. In addition, the scatter plot correlations between revenue and predictor variables popularity and vote count, after log transformations on the x-axis, show an exponential growth pattern with strength indicated by the closeness of points.They suggest increasing popularity and vote counts on movies can predict higher revenue

```python
#Explore the variable genre by revenue proportion
#Identify which genres capture a large percentage of revenue in proportion to total revenue
query = df.groupby(['genre']).revenue.sum()
total_revenue = df.revenue.sum()
genre_df = (query / total_revenue).reset_index().sort_values(by='revenue', ascending=False)

fig = plt.figure(figsize=[10,10])
color = sb.color_palette(palette='pastel')[4]
s = sb.barplot(data=genre_df, x='revenue', y='genre', color=color, order=genre_df['genre'])
s.set_title('Genre By Revenue Proportion')
s.set_xlabel('Revenue (%proportion)', fontsize=12)
s.set_ylabel('Genre', fontsize=12);
```

### Genre By Revenue Proportion
> Action, adventure, and comedy genres produce over 15% of total revenue for all movies. This suggests movies in action, adventure, and comedy genres can be strong predictors for high revenue because they capture a larger percentage of revenue proportion compared to those of other genres. On the other hand, movies in TV Movie and Foreign genres produce less than 1% of total revenue for all movies indicating weak predictors for high revenue

```python
# Identify the highest popularity value for each year in the dataset
# Filter the data for the highest popularity movies by year
# Plot the data as a line chart to see which genre(s) are most popular year to year
query = df.iloc[df.groupby('release_year').popularity.idxmax()].reset_index(drop=True)
query = query.groupby(['release_year', 'genre']).popularity.sum().reset_index()
query = query[~query.popularity.isnull()].reset_index(drop=True).sort_values(by='release_year')

grid = sb.FacetGrid(data=query, hue='genre', height=6)
grid = grid.map(plt.errorbar, 'release_year', 'popularity')
grid.fig.suptitle('Most Popular Genres YOY')
grid.set_xlabels('Release Year', fontsize=12)
grid.set_ylabels('Popularity', fontsize=12)
grid.add_legend(fontsize=11);
```

```python
# In addition to the line plot created above, using the same query dataframe plot the variable genres by frequency
# Identify which genre(s) has the most popularity year to year
fig = plt.figure(figsize=[10,8])
color = sb.color_palette(palette='pastel')[5]
order = query.genre.value_counts().index

sb.countplot(data=query, y='genre', color=color, order=order)
plt.xlabel('Frequency', fontsize=12)
plt.ylabel('Genre', fontsize=12)
plt.title('Most Popular Genres YOY By Frequency', fontsize=12);
```

### Most Popular Genres Year To Year
> The most popular genre year to year is adventure as shown in the line plot and frequency distribution. The action genre is the second most popular genre year to year
